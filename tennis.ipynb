```python
import seaborn
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import csv
```

```python
data = pd.read_csv('./tennis_stats.csv', sep=',')
print(data.head(0))

```

```python
#Exploratory analysis.
plt.scatter(x = data.Wins, y = data.Winnings)
plt.xlabel('Number of Matches won/ Year.')
plt.ylabel('Winning amount $/Year')
plt.title('Winer and Money Won')
plt.show()
plt.clf()
```

# The above visualization shows that winning amount is highly positively correlated to how many matches a player wins in a year.

```python
plt.scatter(x = data.Aces, y = data.Wins)
plt.xlabel('Number of Ace Served/Year')
plt.ylabel('Number of Matches won/Year.')
plt.title('Ace Serve vs Matches Won')
plt.show()
plt.clf()
```

```python
plt.scatter(x = data.TotalServicePointsWon, y = data.Wins)
plt.xlabel('TotalServicePointsWon/Year')
plt.ylabel('Number of Matches won/Year.')
plt.title('Service Poits vs Matches Won')
plt.show()
plt.clf()
```

```python
features = data[['Wins']]
outcome = data[['Winnings']]
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
                
model = LinearRegression()
model.fit(features_train,outcome_train)

model.score(features_test,outcome_test)

prediction = model.predict(features_test)
plt.scatter(outcome_test, prediction, alpha=0.4)
```

```python
features2 = data[['Aces']]
outcome2 = data[['Wins']]
features2_train, features2_test, outcome2_train, outcome2_test = train_test_split(features2, outcome2, train_size = 0.8)
                
model2 = LinearRegression()
model2.fit(features2_train,outcome2_train)

model2.score(features2_test,outcome2_test)

prediction2 = model2.predict(features2_test)
plt.scatter(outcome2_test, prediction2, alpha=0.4)
```

```python
features = data[['TotalServicePointsWon']]
outcome = data[['Wins']]
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
                
model = LinearRegression()
model.fit(features_train,outcome_train)

model.score(features_test,outcome_test)

prediction = model.predict(features_test)
plt.scatter(outcome_test, prediction, alpha=0.4)
```

```python
#Multiple regression analysis.
features = data[['Wins',"Aces"]]
outcome = data[['Winnings']]
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
                
model = LinearRegression()
model.fit(features_train,outcome_train)

model.score(features_test,outcome_test)

prediction = model.predict(features_test)
plt.scatter(outcome_test, prediction, alpha=0.4)
```
